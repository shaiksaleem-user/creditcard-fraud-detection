```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\Saleem\Downloads\creditcard.csv")
print(df.head())
```

```python
# Check the shape of the data
print(f"Dataset shape: {df.shape}")

# Check for missing values
print(f"Missing values:\n{df.isnull().sum()}")

# Summary statistics
print(df.describe())
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()  # Typically, credit card datasets are clean
```

```python
df.info()  # Check columns, data types, non-null counts
df.describe()  # Statistical summary (mean, std, min, max)
```

```python
df['Class'].value_counts(normalize=True)  # Fraud vs. Non-Fraud ratio
sns.countplot(x='Class', data=df)  # Visualize imbalance
```

```python
fraud = df[df['Class'] == 1]['Amount']
non_fraud = df[df['Class'] == 0]['Amount']
print(f"Fraud Amount Stats:\n{fraud.describe()}")
print(f"Non-Fraud Amount Stats:\n{non_fraud.describe()}")
```

```python
corr = df.corr()
plt.figure(figsize=(18,8))
sns.heatmap(corr, cmap='coolwarm', annot=True,fmt='.2f',linewidths=0.5)
```
